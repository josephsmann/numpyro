```python
import numpy as np
import pandas as pd
import altair as alt
import holoviews as hv
```

```python
entropy = lambda a: -(a * np.log(a)).sum()
```

```python
hilf = lambda a: (a**2).sum()
```

```python
_xs = np.linspace(1,10,100)
xs = _xs / _xs.sum()
```

```python
_df = pd.DataFrame({'ent': entropy(xs), 'x': xs, 'hilf': hilf(xs) })
```

```python
base = alt.Chart(_df).mark_point().encode(
    x= 'x:Q')
base.encode(y='ent') + base.encode(y='hilf')
```
